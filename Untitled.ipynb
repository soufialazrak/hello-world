On souponne qu'une personne se connecte en dehors des heures d'ouverture des bureaux (8h-19h), peut-tre depuis un poste distant.

2) Utilisez un script pour retrouver l'identifiant de cette personne ainsi que l'ip  la laquelle elle se connectait


```python
# Extraction des mauvaises dates 
suspet=[]
for i in conx:
    date = i.split(";")[-1]
    full_hour = date.split(" ")[-1]
    hora = full_hour.split(":")[0]
    if int(hora) < 8 or int(hora) > 19 :
            suspet.append(i.split(";")[1])
            listsuspet = set(suspet)
print(listsuspet)        
```

```python

# Enregistrement des suspets dans le fichier suspet.txt
with open('suspet.txt', 'w') as f5:
       for row in listsuspet :
            f5.write(row +"\n")
```

```python
#afficher le contenu du fichier suspet.txt crer
f6 = open('suspet.txt', 'r')

#stockage du contenu dans la 2me variable warn
utils = f6.read()

print(utils)

f6.close()
```

4) A l'aide de cette liste, relevez dans le fichier connexion.log tous les utilisateurs qui se sont connects sur une de ces ip, on produira un fichier suspect.txt avec une ligne par utilisateur et le nombre de fois qu'il s'est connect  une ip interdite

```python
suspet2={}
for i in conx:
    ip = i.split(";")[0]+"\n"
    
    user_name =  i.split(";")[1]
    if ip in warn:
        if user_name in suspet2.keys() :                
                suspet2[user_name] += 1
        else :
                suspet2[user_name] = 1    
print(suspet2)                
```

```python
# Enregistrement des suspets dans le fichier suspet.txt
with open('suspet.txt', 'w') as f5:
       for i,j in suspet2.items() :
            f5.write(f"{i} : {j}\n")
            
           
```
